### **<span style="color:green">Required Packages</span>**  

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### **<span style="color:green">Loading the dataset</span>**  

```python
df = pd.read_csv('train.csv')
```

### **<span style="color:cyan">Total purchases by gender</span>**  

```python
gender_purchase = df.groupby('Gender')['Purchase'].sum().reset_index()
sns.barplot(x='Gender', y='Purchase', data=gender_purchase, palette='Set2')
plt.title('Total Purchases by Gender', fontsize=14)
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Total Purchase Amount', fontsize=12)
plt.tight_layout()
plt.show()
```

### **<span style="color:cyan">Purchase distribution by age group</span>**  

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Age', y='Purchase', data=df, palette='Blues')
plt.title('Purchase Distribution by Age Group', fontsize=14)
plt.xlabel('Age Group', fontsize=12)
plt.ylabel('Purchase Amount', fontsize=12)
plt.tight_layout()
plt.show()
```

### **<span style="color:green">Correlation heatmap</span>**  

```python
plt.figure(figsize=(10, 6))
numerical_cols = ['Purchase', 'Product_Category_1', 'Product_Category_2', 'Product_Category_3']
df_cleaned = df[numerical_cols].dropna()  # Drop rows with missing values
correlation_matrix = df_cleaned.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', cbar_kws={'shrink': 0.8})
plt.title('Correlation Heatmap of Product Categories and Purchase', fontsize=14)
plt.tight_layout()
plt.show()
```

### **<span style="color:green">Customer count by city category</span>**  

```python
sns.countplot(x='City_Category', data=df, palette='pastel', order=df['City_Category'].value_counts().index)
plt.title('Customer Count by City Category', fontsize=14)
plt.xlabel('City Category', fontsize=12)
plt.ylabel('Number of Customers', fontsize=12)
plt.tight_layout()
plt.show()
```
